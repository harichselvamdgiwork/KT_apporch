```python
import os
import re
import csv
from ocr_tamil.ocr import OCR

# Initialize the OCR object
ocr = OCR(detect=True)

# Function to process images and save results in each subfolder
def process_and_save_results(root_folder, output_folder):
    # Iterate over all subfolders in the root folder
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            # Get a list of image files in the subfolder
            image_files = [f for f in os.listdir(folder_path) if f.endswith('.png') or f.endswith('.jpg')]
            total_images = len(image_files)
            print(f"Total number of images in '{folder_name}': {total_images}")

            # Create a CSV file to store the results in the subfolder
            csv_file_path = os.path.join(output_folder, f"{folder_name}_age_gender_results.csv")
            with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(["Filename", "Age", "Gender"])

                # Iterate over all the files in the subfolder
                for i, filename in enumerate(image_files, start=1):
                    # Construct the full path to the image file
                    image_path = os.path.join(folder_path, filename)

                    # Predict text from the image
                    texts = ocr.predict(image_path)

                    # Join the predicted text
                    text = ' '.join(texts[0])

                    # Extract ages above 18 from the text
                    age_pattern = r'(?:வயது\D*(\d{2})|(\d{2})\D*பாலினம்)'
                    age_matches = re.findall(age_pattern, text)
                    ages_above_18 = [age for match in age_matches for age in match if age and int(age) >= 18]

                    # Check for specific words to determine gender
                    gender = None
                    if "பெண்" in text:
                        gender = "female"
                    elif "ஆண்" in text:
                        gender = "male"
                    elif "மூன்றாம் பாலினம்" in text:
                        gender = "transgender"
                    else:
                        gender = "unknown"

                    # Write the results to the CSV file
                    writer.writerow([filename, ', '.join(ages_above_18), gender])
                    
                # Close the CSV file
                print(f"CSV file saved for '{folder_name}'.")

# Root folder containing subfolders with images
root_folder = "/kaggle/working/data33"

# Output folder for CSV files
output_folder = "/kaggle/working/csvfiles33"

# Process images and save results in each subfolder
process_and_save_results(root_folder, output_folder)
```
